```python
import re
from nltk.tokenize import sent_tokenize
from nltk.corpus import stopwords

# Функция для предварительной обработки текста
def preprocess_text(text):
    # Приведение текста к нижнему регистру
    text = text.lower()
    # Удаление пунктуации
    text = re.sub(r'[^\w\s]', '', text)
    # Токенизация на предложения
    sentences = sent_tokenize(text)
    return sentences

# Пример обработки одного из текстов 
sample_text = articles_for_term_extraction['articles'].iloc[0]  
processed_sentences = preprocess_text(sample_text)

#print("Обработанные предложения:")
print(processed_sentences[:5])
```

```python
# Применение обработки ко всему датасету
articles_for_term_extraction['processed_text'] = articles_for_term_extraction['articles'].apply(preprocess_text)

print("Обработанные тексты:")
print(articles_for_term_extraction['processed_text'].head())
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Функция для визуализации в 2D пространстве
def visualize_terms(term_vecs, term_labels, title):
    pca = PCA(n_components=2)
    reduced_vecs = pca.fit_transform(term_vecs)
    
    plt.figure(figsize=(10, 8))
    plt.scatter(reduced_vecs[:, 0], reduced_vecs[:, 1], c='blue', edgecolor='k', s=50)
    
    for i, label in enumerate(term_labels):
        plt.annotate(label, (reduced_vecs[i, 0], reduced_vecs[i, 1]), fontsize=12)
    
    plt.title(title)
    plt.xlabel('PCA Component 1')
    plt.ylabel('PCA Component 2')
    plt.grid(True)
    plt.show()

# Визуализация терминов для CE и SC
visualize_terms(textrank_CE_vecs, ds_CE_textrank_terms['term'], 'TextRank Terms - CE')
visualize_terms(tfidf_CE_vecs, ds_CE_tfidf_terms['term'], 'TF-IDF Terms - CE')
visualize_terms(textrank_SC_vecs, ds_SC_textrank_terms['term'], 'TextRank Terms - SC')
visualize_terms(tfidf_SC_vecs, ds_SC_tfidf_terms['term'], 'TF-IDF Terms - SC')
```

# Сравнение терминов полученных двумя способами. Оценка двух способов. Подсчет **Entropy** and **Variance**

```python
import pandas as pd
import numpy as np
from scipy.stats import entropy
from sklearn.feature_extraction.text import CountVectorizer

# Функция для вычисления дисперсии и энтропии
def compute_variance_entropy(term_freq):
    term_var = np.var(term_freq, axis=0)
    
    def safe_entropy(freq):
        if np.sum(freq) == 0:
            return 0  # Устанавливаем энтропию равной нулю для нулевых частот
        return entropy(freq)
    
    term_entropy = np.apply_along_axis(safe_entropy, axis=0, arr=term_freq.T)
    return term_var, term_entropy

# Удаление или заполнение NaN значений
ds_CE_textrank_terms.dropna(subset=['context'], inplace=True)
ds_SC_textrank_terms.dropna(subset=['context'], inplace=True)
ds_CE_tfidf_terms.dropna(subset=['context_sent'], inplace=True)
ds_SC_tfidf_terms.dropna(subset=['context_sent'], inplace=True)

# Получение списка документов из 'context'
docs_CE_textrank = ds_CE_textrank_terms['context'].tolist()
docs_SC_textrank = ds_SC_textrank_terms['context'].tolist()
docs_CE_tfidf = ds_CE_tfidf_terms['context_sent'].tolist()
docs_SC_tfidf = ds_SC_tfidf_terms['context_sent'].tolist()

# Преобразование документов в строки и приведение к нижнему регистру
docs_CE_textrank = [str(doc).lower() for doc in docs_CE_textrank]
docs_SC_textrank = [str(doc).lower() for doc in docs_SC_textrank]
docs_CE_tfidf = [str(doc).lower() for doc in docs_CE_tfidf]
docs_SC_tfidf = [str(doc).lower() for doc in docs_SC_tfidf]

# Создание списка всех уникальных терминов
all_terms = list(set(ds_CE_textrank_terms['term']).union(
                 set(ds_SC_textrank_terms['term'])).union(
                 set(ds_CE_tfidf_terms['term'])).union(
                 set(ds_SC_tfidf_terms['term'])))

# Создание кастомного векторизатора, включающего все термины
class CustomVectorizer(CountVectorizer):
    def build_analyzer(self):
        analyzer = super().build_analyzer()
        return lambda doc: [w for w in analyzer(doc) if w in all_terms]

vectorizer = CustomVectorizer(ngram_range=(1, 3), vocabulary=all_terms)

# Векторизация документов
X_CE_textrank = vectorizer.fit_transform(docs_CE_textrank)
X_SC_textrank = vectorizer.fit_transform(docs_SC_textrank)
X_CE_tfidf = vectorizer.fit_transform(docs_CE_tfidf)
X_SC_tfidf = vectorizer.fit_transform(docs_SC_tfidf)

# Получение термов из векторизаторов
terms = vectorizer.get_feature_names_out()

# Функция для безопасного обновления DataFrame
def safe_update_df(df, term_list, variance, entropy, term_col='term'):
    term_to_variance = dict(zip(term_list, variance))
    term_to_entropy = dict(zip(term_list, entropy))
    
    df['variance'] = df[term_col].map(term_to_variance)
    df['entropy'] = df[term_col].map(term_to_entropy)

# Вычисление дисперсии и энтропии для каждого терма
variance_CE_textrank, entropy_CE_textrank = compute_variance_entropy(X_CE_textrank.toarray())
variance_SC_textrank, entropy_SC_textrank = compute_variance_entropy(X_SC_textrank.toarray())
variance_CE_tfidf, entropy_CE_tfidf = compute_variance_entropy(X_CE_tfidf.toarray())
variance_SC_tfidf, entropy_SC_tfidf = compute_variance_entropy(X_SC_tfidf.toarray())

# Обновление DataFrame для CE TextRank
safe_update_df(ds_CE_textrank_terms, terms, variance_CE_textrank, entropy_CE_textrank)

# Обновление DataFrame для SC TextRank
safe_update_df(ds_SC_textrank_terms, terms, variance_SC_textrank, entropy_SC_textrank)

# Обновление DataFrame для CE TF-IDF
safe_update_df(ds_CE_tfidf_terms, terms, variance_CE_tfidf, entropy_CE_tfidf)

# Обновление DataFrame для SC TF-IDF
safe_update_df(ds_SC_tfidf_terms, terms, variance_SC_tfidf, entropy_SC_tfidf)

# Создание списков уникальных и общих терминов
unique_textrank = set(ds_CE_textrank_terms['term']).union(set(ds_SC_textrank_terms['term'])) - set(ds_CE_tfidf_terms['term']).union(set(ds_SC_tfidf_terms['term']))
unique_tfidf = set(ds_CE_tfidf_terms['term']).union(set(ds_SC_tfidf_terms['term'])) - set(ds_CE_textrank_terms['term']).union(set(ds_SC_textrank_terms['term']))
common_terms = set(ds_CE_textrank_terms['term']).union(set(ds_SC_textrank_terms['term'])).intersection(set(ds_CE_tfidf_terms['term']).union(set(ds_SC_tfidf_terms['term'])))

# Замена NaN значений на 0 или другие значения по вашему усмотрению
ds_CE_textrank_terms['entropy'] = ds_CE_textrank_terms['entropy'].fillna(0)
ds_SC_textrank_terms['entropy'] = ds_SC_textrank_terms['entropy'].fillna(0)
ds_CE_tfidf_terms['entropy'] = ds_CE_tfidf_terms['entropy'].fillna(0)
ds_SC_tfidf_terms['entropy'] = ds_SC_tfidf_terms['entropy'].fillna(0)

# Проверка обновленных данных
print("Проверка на наличие NaN значений в ds_CE_textrank_terms после обновления:")
print(ds_CE_textrank_terms.isna().sum())

print("Проверка на наличие NaN значений в ds_SC_textrank_terms после обновления:")
print(ds_SC_textrank_terms.isna().sum())

print("Проверка на наличие NaN значений в ds_CE_tfidf_terms после обновления:")
print(ds_CE_tfidf_terms.isna().sum())

print("Проверка на наличие NaN значений в ds_SC_tfidf_terms после обновления:")
print(ds_SC_tfidf_terms.isna().sum())

# Печать количества уникальных и общих терминов
print("Уникальные термины TextRank (не найденные TF-IDF):", len(unique_textrank))
print("Уникальные термины TF-IDF (не найденные TextRank):", len(unique_tfidf))
print("Общие термины TextRank и TF-IDF:", len(common_terms))
```

# Поиск уникальных и общих терминов для SC и CE (анализ двух способов по-отдельности)

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer

# Удаление или заполнение NaN значений
ds_CE_textrank_terms.dropna(subset=['context'], inplace=True)
ds_SC_textrank_terms.dropna(subset=['context'], inplace=True)
ds_CE_tfidf_terms.dropna(subset=['context_sent'], inplace=True)
ds_SC_tfidf_terms.dropna(subset=['context_sent'], inplace=True)

# Получение списка документов из 'context'
docs_CE_textrank = ds_CE_textrank_terms['context'].tolist()
docs_SC_textrank = ds_SC_textrank_terms['context'].tolist()
docs_CE_tfidf = ds_CE_tfidf_terms['context_sent'].tolist()
docs_SC_tfidf = ds_SC_tfidf_terms['context_sent'].tolist()

# Преобразование документов в строки и приведение к нижнему регистру
docs_CE_textrank = [str(doc).lower() for doc in docs_CE_textrank]
docs_SC_textrank = [str(doc).lower() for doc in docs_SC_textrank]
docs_CE_tfidf = [str(doc).lower() for doc in docs_CE_tfidf]
docs_SC_tfidf = [str(doc).lower() for doc in docs_SC_tfidf]

# Создание списка всех уникальных терминов
all_terms_textrank = list(set(ds_CE_textrank_terms['term']).union(set(ds_SC_textrank_terms['term'])))
all_terms_tfidf = list(set(ds_CE_tfidf_terms['term']).union(set(ds_SC_tfidf_terms['term'])))

# Создание кастомного векторизатора, включающего все термины
class CustomVectorizer(CountVectorizer):
    def __init__(self, vocabulary):
        super().__init__(ngram_range=(1, 3), vocabulary=vocabulary)

    def build_analyzer(self):
        analyzer = super().build_analyzer()
        return lambda doc: [w for w in analyzer(doc) if w in self.vocabulary]

vectorizer_textrank = CustomVectorizer(vocabulary=all_terms_textrank)
vectorizer_tfidf = CustomVectorizer(vocabulary=all_terms_tfidf)

# Векторизация документов
X_CE_textrank = vectorizer_textrank.fit_transform(docs_CE_textrank)
X_SC_textrank = vectorizer_textrank.fit_transform(docs_SC_textrank)
X_CE_tfidf = vectorizer_tfidf.fit_transform(docs_CE_tfidf)
X_SC_tfidf = vectorizer_tfidf.fit_transform(docs_SC_tfidf)

# Получение термов из векторизаторов
terms_textrank = vectorizer_textrank.get_feature_names_out()
terms_tfidf = vectorizer_tfidf.get_feature_names_out()

# Выделение уникальных и общих терминов
unique_CE_textrank = set(ds_CE_textrank_terms['term']) - set(ds_SC_textrank_terms['term'])
unique_SC_textrank = set(ds_SC_textrank_terms['term']) - set(ds_CE_textrank_terms['term'])
common_textrank = set(ds_CE_textrank_terms['term']).intersection(set(ds_SC_textrank_terms['term']))

unique_CE_tfidf = set(ds_CE_tfidf_terms['term']) - set(ds_SC_tfidf_terms['term'])
unique_SC_tfidf = set(ds_SC_tfidf_terms['term']) - set(ds_CE_tfidf_terms['term'])
common_tfidf = set(ds_CE_tfidf_terms['term']).intersection(set(ds_SC_tfidf_terms['term']))

# Печать результатов
print("TextRank метод:")
print("Уникальные термины CE TextRank:", len(unique_CE_textrank), ":", unique_CE_textrank)
print("Уникальные термины SC TextRank:", len(unique_SC_textrank), ":", unique_SC_textrank)
print("Общие термины TextRank:", len(common_textrank), ":", common_textrank)

print("\nTF-IDF метод:")
print("Уникальные термины CE TF-IDF:", len(unique_CE_tfidf), ":", unique_CE_tfidf)
print("Уникальные термины SC TF-IDF:", len(unique_SC_tfidf), ":", unique_SC_tfidf)
print("Общие термины TF-IDF:", len(common_tfidf), ":", common_tfidf)
```

# Для лучшей визуализации убрали outliers

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Функция для фильтрации выбросов
def filter_outliers(df, column, threshold=3):
    mean = df[column].mean()
    std = df[column].std()
    return df[np.abs(df[column] - mean) <= threshold * std]

# Фильтрация выбросов для каждого DataFrame
ds_CE_textrank_terms_filtered = filter_outliers(ds_CE_textrank_terms, 'variance')
ds_SC_textrank_terms_filtered = filter_outliers(ds_SC_textrank_terms, 'variance')
ds_CE_tfidf_terms_filtered = filter_outliers(ds_CE_tfidf_terms, 'variance')
ds_SC_tfidf_terms_filtered = filter_outliers(ds_SC_tfidf_terms, 'variance')

ds_CE_textrank_terms_filtered = filter_outliers(ds_CE_textrank_terms_filtered, 'entropy')
ds_SC_textrank_terms_filtered = filter_outliers(ds_SC_textrank_terms_filtered, 'entropy')
ds_CE_tfidf_terms_filtered = filter_outliers(ds_CE_tfidf_terms_filtered, 'entropy')
ds_SC_tfidf_terms_filtered = filter_outliers(ds_SC_tfidf_terms_filtered, 'entropy')

# Подготовка данных для визуализации
ds_CE_textrank_terms_filtered['method'] = 'TextRank'
ds_CE_textrank_terms_filtered['area'] = 'CE'
ds_SC_textrank_terms_filtered['method'] = 'TextRank'
ds_SC_textrank_terms_filtered['area'] = 'SC'
ds_CE_tfidf_terms_filtered['method'] = 'TF-IDF'
ds_CE_tfidf_terms_filtered['area'] = 'CE'
ds_SC_tfidf_terms_filtered['method'] = 'TF-IDF'
ds_SC_tfidf_terms_filtered['area'] = 'SC'

df_all_filtered = pd.concat([ds_CE_textrank_terms_filtered, ds_SC_textrank_terms_filtered, ds_CE_tfidf_terms_filtered, ds_SC_tfidf_terms_filtered])

# Визуализация распределения дисперсии и энтропии
plt.figure(figsize=(14, 10))

# Гистограммы для дисперсии
plt.subplot(2, 2, 1)
sns.histplot(df_all_filtered[df_all_filtered['method'] == 'TextRank'], x='variance', hue='area', bins=30, kde=True)
plt.title('Variance Distribution - TextRank')
plt.ylim(0, 50)  # Устанавливаем пределы оси Y для лучшего масштабирования

plt.subplot(2, 2, 2)
sns.histplot(df_all_filtered[df_all_filtered['method'] == 'TF-IDF'], x='variance', hue='area', bins=30, kde=True)
plt.title('Variance Distribution - TF-IDF')
plt.ylim(0, 50)  # Устанавливаем пределы оси Y для лучшего масштабирования

# Гистограммы для энтропии
plt.subplot(2, 2, 3)
sns.histplot(df_all_filtered[df_all_filtered['method'] == 'TextRank'], x='entropy', hue='area', bins=30, kde=True)
plt.title('Entropy Distribution - TextRank')
plt.ylim(0, 50)  # Устанавливаем пределы оси Y для лучшего масштабирования

plt.subplot(2, 2, 4)
sns.histplot(df_all_filtered[df_all_filtered['method'] == 'TF-IDF'], x='entropy', hue='area', bins=30, kde=True)
plt.title('Entropy Distribution - TF-IDF')
plt.ylim(0, 50)  # Устанавливаем пределы оси Y для лучшего масштабирования

plt.tight_layout()
plt.show()
```
